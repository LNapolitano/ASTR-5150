```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.special
```

```python
#Problem 3, plotting Poisson and Gaussian Distributions

cen=4 #center of distributions
dev=2 #standard dev of distributions

def Gaussian(x,mu,sig): #x,mean,st.dev #PGaussian Distribution
    return(np.exp(-(x-mu)**2/(2*sig**2))/(sig*np.sqrt(2*np.pi)))

def Poisson(x,mu):
    return(np.exp(-mu)*(mu**x)/(scipy.special.factorial(x)))


x_ran=np.arange(cen-10*dev,cen+10*dev,0.01)

plt.plot(np.arange(cen-10*dev,cen+10*dev,0.01),Gaussian(x_ran,cen,dev))  
plt.plot(np.arange(cen-10*dev,cen+10*dev,0.01),Poisson(x_ran,cen))
plt.xlabel("X value")
plt.ylabel("Probability")
plt.show()

cen=25
dev=5

x_ran=np.arange(cen-10*dev,cen+10*dev,0.01)

plt.plot(np.arange(cen-10*dev,cen+10*dev,0.01),Gaussian(x_ran,cen,dev))  
plt.plot(np.arange(cen-10*dev,cen+10*dev,0.01),Poisson(x_ran,cen))
plt.xlabel("X value")
plt.ylabel("Probability")
plt.show()
```

```python
#Problem 4 - Integrating Distributions

cen=0
dev=10

x_ran=np.arange(cen-10*dev,cen+10*dev,0.01)
plt.plot(np.arange(cen-10*dev,cen+10*dev,0.01),Gaussian(x_ran,cen,dev)) 
plt.show()

full=np.arange(cen-100*dev,cen+100*dev,0.01)

sig1=np.arange(cen+1*dev,cen+100*dev,0.01)
sig3=np.arange(cen+3*dev,cen+100*dev,0.01)
sig5=np.arange(cen+5*dev,cen+100*dev,0.01)

print((np.trapz(Gaussian(full,cen,dev))))

print(2*np.trapz(Gaussian(sig1,cen,dev)))
print(2*np.trapz(Gaussian(sig3,cen,dev))) 
print(2*np.trapz(Gaussian(sig5,cen,dev)))
```
